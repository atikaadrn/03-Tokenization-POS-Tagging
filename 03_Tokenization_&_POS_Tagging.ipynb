# spaCy

```python
import spacy
import pandas as pd

# Pastikan model sudah diunduh sebelumnya: !python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")

# --- MEMENUHI METRIK: 10 Kalimat dengan 3 Jenis Berbeda ---
kalimat_list = [
    # 1. Deklaratif (Pernyataan)
    "The algorithm analyzes the text sentiment accurately.",
    "Lexical normalization improves the baseline performance.",
    "They build an interactive dashboard.",
    "The dictionary contains duplicate words.",

    # 2. Interrogative (Pertanyaan)
    "How does the model classify a neutral emotion?",
    "Are we using a rule-based approach today?",
    "What is the final accuracy of this pipeline?",

    # 3. Kompleks (Lebih dari satu klausa)
    "Although the dataset is large, the preprocessing phase is fast.",
    "The researcher, who studies data science, published a great article.",
    "If the rating aligns with the lexicon, we consider it congruent."
]

# --- PROSES TOKENISASI & POS TAGGING ---
hasil_ekstraksi = []

for i, teks in enumerate(kalimat_list, 1):
    doc = nlp(teks)
    for token in doc:
        pos = token.pos_

        # Filter untuk memberi label khusus pada target: Noun, Verb, Adj
        if pos in ["NOUN", "PROPN"]:
            label_target = "Noun (Kata Benda)"
        elif pos in ["VERB", "AUX"]:
            label_target = "Verb (Kata Kerja)"
        elif pos == "ADJ":
            label_target = "Adj (Kata Sifat)"
        else:
            label_target = "Lainnya" # Mengabaikan kata depan, tanda baca, dll

        # Menyimpan ke dalam list (Hanya ambil yang Noun, Verb, Adj agar fokus pada Goal)
        if label_target != "Lainnya":
            hasil_ekstraksi.append({
                "ID Kalimat": f"Kalimat {i}",
                "Token": token.text,
                "POS Tag": pos,
                "Kategori Target": label_target,
                "Keterangan/Dokumentasi": spacy.explain(token.tag_)
            })

# --- MEMENUHI METRIK: Tampilkan dalam tabel yang rapi ---
df_hasil = pd.DataFrame(hasil_ekstraksi)

# Menampilkan 15 baris pertama dari tabel
print(df_hasil.head(15).to_markdown(index=False))
```

# NLTK

```python
import nltk
import pandas as pd

# Pastikan resources NLTK terunduh
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab') # Added to address LookupError
nltk.download('averaged_perceptron_tagger_eng') # Added to address new LookupError

# --- DATA: 10 Kalimat dengan 3 Jenis Berbeda (Sama seperti spaCy) ---
kalimat_list = [
    # Deklaratif
    "The algorithm analyzes the text sentiment accurately.",
    "Lexical normalization improves the baseline performance.",
    "They build an interactive dashboard.",
    "The dictionary contains duplicate words.",
    # Interogatif
    "How does the model classify a neutral emotion?",
    "Are we using a rule-based approach today?",
    "What is the final accuracy of this pipeline?",
    # Kompleks
    "Although the dataset is large, the preprocessing phase is fast.",
    "The researcher, who studies data science, published a great article.",
    "If the rating aligns with the lexicon, we consider it congruent."
]

hasil_nltk = []

# --- PROSES MENGGUNAKAN NLTK ---
for i, teks in enumerate(kalimat_list, 1):
    # 1. Tokenisasi dengan NLTK
    tokens = nltk.word_tokenize(teks);

    # 2. POS Tagging dengan NLTK
    pos_tags = nltk.pos_tag(tokens)

    for token, tag in pos_tags:
        # 3. Mapping Penn Treebank Tags ke target Anda (Noun, Verb, Adj)
        if tag.startswith('NN'):
            # NN, NNS, NNP, NNPS adalah kelompok Noun
            label_target = "Noun (Kata Benda)"
        elif tag.startswith('VB') or tag == 'MD':
            # VB, VBD, VBG, VBN, VBP, VBZ, MD adalah kelompok Verb/Auxiliary
            label_target = "Verb (Kata Kerja)"
        elif tag.startswith('JJ'):
            # JJ, JJR, JJS adalah kelompok Adjective
            label_target = "Adj (Kata Sifat)"
        else:
            label_target = "Lainnya"

        if label_target != "Lainnya":
            hasil_nltk.append({
                "ID Kalimat": f"Kalimat {i}",
                "Token": token,
                "NLTK Tag": tag,
                "Kategori Target": label_target
            })

# --- CETAK TABEL ---
df_nltk = pd.DataFrame(hasil_nltk)
print(df_nltk.head(15).to_markdown(index=False))
```

kenapa ada beberapa token tidak masuk ("the", "accurately" karena secara spesifik dirancang untuk hanya menyertakan token yang dikategorikan sebagai 'Noun', 'Verb', atau 'Adj' dalam tabel akhir.

### Perbandingan POS Tagging vs NER
Sel jalankan di bawah ini menunjukkan bagaimana spaCy membedakan kategori kata secara gramatikal (POS) dan entitas nama (NER).

```python
import spacy

nlp = spacy.load("en_core_web_sm")
teks_contoh = "Apple is looking at buying a U.K. startup for $1 billion in London."
doc = nlp(teks_contoh)

# --- 1. Hasil POS Tagging ---
print("--- POS TAGGING (Gramatikal) ---")
for token in doc[:5]: # Tampilkan 5 kata pertama
    print(f"{token.text:10} -> {token.pos_:7} ({spacy.explain(token.pos_)})")

print("\n--- NER (Entitas Nama) ---")
# --- 2. Hasil NER ---
for ent in doc.ents:
    print(f"{ent.text:15} -> {ent.label_:10} ({spacy.explain(ent.label_)})")
```

### Menampilkan Coarse-grained (pos_) vs Fine-grained (tag_) Tags
Kode di bawah ini menunjukkan daftar lengkap kategori yang tersedia dan bagaimana keduanya dipasangkan.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying a U.K. startup.")

print(f"{'Teks':<10} | {'Coarse (pos_)':<15} | {'Fine (tag_)':<15} | {'Penjelasan Detail':<20}")
print("-" * 70)
for token in doc:
    print(f"{token.text:<10} | {token.pos_:<15} | {token.tag_:<15} | {spacy.explain(token.tag_)}")
```

### Perbedaan Visual: Tokenisasi vs POS Tagging

- **Tokenisasi**: `["Saya", "belajar", "Python"]`
- **POS Tagging**: `[ ("Saya", "PRON"), ("belajar", "VERB"), ("Python", "PROPN") ]`

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Data science is fascinating.")

print(f"{'Tahap':<15} | {'Hasil Output'}")
print("-" * 40)

# 1. Tokenisasi saja
tokens = [token.text for token in doc]
print(f"{'Tokenisasi':<15} | {tokens}")

# 2. POS Tagging (menambahkan label ke token)
pos_labels = [f"{token.text}({token.pos_})" for token in doc]
print(f"{'POS Tagging':<15} | {pos_labels}")
```
